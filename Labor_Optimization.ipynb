#Given a factory with information:

The factory produces 7 days a week with 3 shifts a day: morning, afternoon and evening.
The factory has the production plan for next week with estimated the workforce daily workforce requirement:
Monday (6,4,3)
Tuesday (5,5,5)
Wednesday (6,4,4)
Thursday (5,5,5)
Friday (5,5,3)
Saturday (6,5,4)
Sunday (5,5,4).
The total workforce is 20 people.
According to the labor rights:
A labor can only work 1 shift a day and each week
A labor must take leave at least 2 days
If a labor works on evening shift today, he/she cannot work on the morning shift tomorrow.
Therefore, as a production planner, how can we assign each labor to each day and each shift to minimize the cost, satisfy the demand but still follow the rights?

```python
# Sets
I = 20
J = 7
K = 3

# Parameter:
D = [
    [6,4,3],    # Monday
    [5,5,5],    # Tuesday
    [6,4,4],    # Wednesday
    [5,5,5],    # Thursday
    [5,5,3],    # Friday
    [6,5,4],    # Saturday
    [5,5,4]     # Sunday
]
```

```python
# Set up  Model
m = gp.Model('Labor_Scheduling')
```

```python
# Decision Variable
x = m.addVars(I,J,K,vtype = gp.GRB.BINARY, name ='x')
```

```python
# Objective Function
m.setObjective(gp.quicksum(x[i,j,k] for i in range(I) for j in range(J) for k in range(K)),gp.GRB.MINIMIZE)
```

```python
# Constraints
## Demand Constraints
# Cách 1:
for j in range(J):
  for k in range(K):
    m.addConstr(gp.quicksum(x[i,j,k] for i in range(I)) == D[j][k], name ='Demand Fullfill')

# Cách 2:
# m.addConstr(gp.quicksum(x[i,j,k] for i in range(I) ) == D[j][k] for j in range(J) for k in range(K))

## Each labor work only 1 shift per day
for i in range(I):
  for j in range(J):
    m.addConstr(gp.quicksum(x[i,j,k] for k in range(K)) <= 1, name ='Each labor work only 1 shift per day' )

## Each labor must leave 2 day per week
for i in range(I):
  m.addConstr(gp.quicksum(x[i,j,k] for j in range(J) for k in range(K))<=5 )

## Labor who work last night, cannot work tomorrow morning --> Each labor i in each 2 days j and j+1
for i in range(I):
  for j in range(J-1):
    m.addConstr(x[i,j,2] + x[i,j+1,0] <= 1 , name ='Labor who work last night, cannot work tomorrow morning')

# Run Optimize
m.optimize()
```

```python
for v in m.getVars():
    print(f"{v.VarName} = {v.X}")
```

```python
import pandas as pd
```

```python
if m.status == gp.GRB.OPTIMAL:
    # Prepare data for the table
    table_data = []

    for i in range(I):
        for j in range(J):
            shifts = [k + 1 for k in range(K) if x[i, j, k].x > 1e-6]
            if shifts:
                for k in shifts:
                    table_data.append({
                        'Staff': i + 1,  # Staff numbers starting from 1
                        'Day': j + 1,    # Days starting from 1
                        'Shift': k,      # Shifts starting from 1
                        'Status': 'Assigned'
                    })
            else:
                # No shifts assigned on this day, so staff is on leave
                table_data.append({
                    'Staff': i + 1,
                    'Day': j + 1,
                    'Shift': 'N/A',
                    'Status': 'Leave'
                })

    # Create DataFrame
    df = pd.DataFrame(table_data)

    # Print the DataFrame
    print(df)

    # Optionally, save to a CSV file
    df.to_csv('staff_assignment.csv', index=False)

else:
    print("No optimal solution found.")
```
